```python
import cv2

#loading face and smile cascades
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
smile_cascade=cv2.CascadeClassifier('haarcascade_smile.xml')

#Opening Camera we give give only 2 value as argument:
# 0 : If we want to run our internal camera
# 1 : If we want to run our external camera
cam = cv2.VideoCapture(0)

#Loop run unit ESC key is not pressed 
while True : 
    # cam.read() return two frames 1 for checking img and second img of last frame 
    #since we don't want 1st one and need only 2nd i use _ for one to avoid unncesarry storage occupation
    _ , img = cam.read()
    
    # Since Haarcascade was trained on grey scale images so first we need to convert our colorful image to black and white 
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    # detectMultiscale will detect our face 1.3,5 are two tuneable parameters 
    # detectMultiscale will return tuple of 4 cordinates i.e (top left x , y , width , height)
    
    face = face_cascade.detectMultiScale(
                                        gray,
                                        1.3,
                                        5
                                        )
    
    
    # We are running loop all over our face to make a rectangle box
    for (fx,fy,fw,fh) in face :
        cv2.rectangle(img,(fx,fy),(fx+fw,fy+fh),(255,0,0),2)
       
    # Since we don't this model to detect our smile outside our face so here we are setting up roi our face area
        roi = gray[fy:fy+fh,fx:fx+fw]
        color_roi = img[fy:fy+fh,fx:fx+fw]
        
        # it will detect wether we are smiling or not in our roi 
        smile = smile_cascade.detectMultiScale( roi,
                                                1.7,
                                                20
                                               )
        smile 
        for (sx,sy,sw,sh) in smile :
            cv2.rectangle(color_roi,(sx,sy),(sx+sw,sy+sh),(0,255,0),2)
                          # doubt   
        for i in smile :
            # will write smilling if we are smiling in more than 1  frame 
            if(len(i)>1):
                cv2.putText(img,"Smiling",(fx,fy-30),cv2.FONT_HERSHEY_SIMPLEX,
                        2,(0,0,255),3,cv2.LINE_AA)
            elif(len(i)<=1):
                cv2.putText(img,"Not Smilling",(fx,fy-30),cv2.FONT_HERSHEY_SIMPLEX,
                    2,(0,0,255),3,cv2.LINE_AA)
    
    # will return the img 
    cv2.imshow('video', img)
    # This loop will run until we press 'ESC' key..
    k = cv2.waitKey(30) & 0xff
    if k == 27: # press 'ESC' to quit
        break

# After termination of loop this will release our camera and destroy all it's window 
cam.release()
cv2.destroyAllWindows()
    
```
